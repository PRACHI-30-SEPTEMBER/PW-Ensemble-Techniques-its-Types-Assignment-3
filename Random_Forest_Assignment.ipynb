# Random Forest Regressor Assignment

Questions and Answers

### Q1. What is Random Forest Regressor?

**Answer:**

Random Forest Regressor is an ensemble machine learning algorithm that builds multiple decision trees and combines their outputs to make more accurate and stable predictions. It uses bagging and random feature selection.

### Q2. How does Random Forest Regressor reduce the risk of overfitting?

**Answer:**

It reduces overfitting by training multiple trees on different subsets of data and features. The averaging of predictions reduces variance and prevents the model from fitting noise.

### Q3. How does Random Forest Regressor aggregate the predictions of multiple decision trees?

**Answer:**

It aggregates predictions by taking the average of outputs from all individual decision trees in the forest.

### Q4. What are the hyperparameters of Random Forest Regressor?

**Answer:**

Important hyperparameters include:
- n_estimators (number of trees)
- max_depth
- min_samples_split
- min_samples_leaf
- max_features
- bootstrap

### Q5. What is the difference between Random Forest Regressor and Decision Tree Regressor?

**Answer:**

Decision Tree uses a single tree and can overfit easily.
Random Forest uses multiple trees and reduces overfitting by averaging results, making it more robust and accurate.

### Q6. What are the advantages and disadvantages of Random Forest Regressor?

**Answer:**

Advantages:
- High accuracy
- Reduces overfitting
- Handles large datasets well

Disadvantages:
- Computationally expensive
- Less interpretable
- Slower prediction time

### Q7. What is the output of Random Forest Regressor?

**Answer:**

It outputs a continuous numerical value which is the average prediction of all decision trees.

### Q8. Can Random Forest Regressor be used for classification tasks?

**Answer:**

No, Random Forest Regressor is specifically for regression. For classification tasks, Random Forest Classifier is used.
